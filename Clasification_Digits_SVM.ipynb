```python
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
import numpy as np
%matplotlib inline
```

```python
digits = datasets.load_digits()
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target)
```

```python
models = [svm.SVC(kernel='linear', C=1.0),
                 svm.LinearSVC(C=1.0),
                 svm.SVC(kernel='rbf', gamma=0.7, C=1.0),
                 svm.SVC(kernel='poly', degree=3, C=1.0)]
```

```python
models = [clf.fit(x_train, y_train) for clf in models]
```

```python
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']
```

```python
n = 1090
for clf, title, ind in zip(models, titles, index):
    pre = clf.predict([digits.data[n]])
    p = clf.score(x_test, y_test)
    plt.figure(1)
    plt.imshow(digits.images[n], cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title("%s - Precisión: %s" % (title, p))
    plt.ylabel("Predicción: %i" % pre)
    plt.show()
```
